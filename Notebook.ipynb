```python
df["con"] = df['title'] + " " + df['genres']
df.head()
```

```python
tv = TfidfVectorizer()
vector = tv.fit_transform(df['con'])
```

```python
similar_score = cosine_similarity(vector)
```

```python
movie = input("Enter movie name :")
```

```python
movie_list = list(df['title'])

close_match = d.get_close_matches(movie, movie_list)
close_match = close_match[0]

print(close_match)
```

```python
index_of_movie = df.loc[df['title'] == close_match].index[0]
```

```python
index_of_movie
```

```python
score_list = list(enumerate(similar_score[index_of_movie]))
```

```python
sorted_score_list = sorted(score_list, key = lambda x:x[1], reverse = True) 
print(sorted_score_list)
```

```python
sorted_score_list[0][0]
```

```python
id = []
for i in range(0, 11):
    x = sorted_score_list[i][0]
    id.append(x)
    
print(id)
```

```python
print('Movies suggested for you : \n')

for i in id:
    dg = df[df['movieId'] == i]['title'].values
    print(dg)
```
